# Min-Max Scaling

```python
import pandas as pd

# Step 1: Read the CSV file
file_path = 'AIRD_Elevation_Joined.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Step 2: Extract the relevant columns
air_days = data['AIRD']
elevation = data['ELEVATION (m)']

# Step 3: Min-max scaling function
def min_max_scaling(series):
    min_val = series.min()
    max_val = series.max()
    scaled = (series - min_val) / (max_val - min_val)
    return scaled

# Apply the scaling
scaled_air_days = min_max_scaling(air_days)
scaled_elevation = min_max_scaling(elevation)

# Step 4: Calculate the final score
data['Final Score'] = scaled_air_days + scaled_elevation

# Optional: Save the results to a new CSV file
data.to_csv('scaled_results.csv', index=False)

print(data[['AIRD', 'ELEVATION (m)', 'Final Score']])
```

```python
import pandas as pd

# Load the data with both 'AIRD' and 'ELEVATION (m)' columns
file_path = 'AIRD_Elevation_Joined.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Step 1: Extract the relevant columns
elevation = data['ELEVATION (m)']

# Step 2: Calculate mean and standard deviation
mean_elevation = elevation.mean()
std_dev_elevation = elevation.std()

# Step 3: Define the ideal range
ideal_min = 1500
ideal_max = 3000

# Step 4: Calculate the scores for the elevation based on z-scores
def elevation_score(elev):
    if elev < ideal_min or elev > ideal_max:
        # Calculate z-score
        z_score = (elev - mean_elevation) / std_dev_elevation
        # Penalize the score based on how far the elevation is from the mean
        return 1 - abs(z_score)  # This will give lower scores for extremes
    else:
        # Scale positively within the ideal range
        return (elev - ideal_min) / (ideal_max - ideal_min)  # Scale to 0-1

# Apply the scoring function
data['Elevation Score'] = elevation.apply(elevation_score)

# Step 5: Min-max scaling for 'AIRD' column
def min_max_scaling(series):
    min_val = series.min()
    max_val = series.max()
    scaled = (series - min_val) / (max_val - min_val)
    return scaled

# Apply the scaling to 'AIRD'
scaled_air_days = min_max_scaling(data['AIRD'])

# Step 6: Calculate the final score
data['Final Score'] = scaled_air_days + data['Elevation Score']

# Optional: Get the top 10 places based on the final score
top_10_places = data.nlargest(10, 'Final Score')[['STATION NAME', 'STATE', 'Final Score', 'ELEVATION (m)', 'AIRD']]

# Print the results
print(top_10_places)
```

```python
import pandas as pd

# Load the data with both 'AIRD' and 'ELEVATION (m)' columns
file_path = 'AIRD_Elevation_Joined.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Step 1: Extract the relevant columns
elevation = data['ELEVATION (m)']

# Step 2: Define the ideal range
ideal_min = 1500
ideal_max = 3000

# Step 3: Calculate the scores for the elevation
def elevation_score(elev):
    if ideal_min <= elev <= ideal_max:
        return 1  # Maximum score for ideal range
    else:
        # Linear drop-off scoring outside the ideal range
        if elev < ideal_min:
            # Calculate penalty based on distance from ideal_min
            penalty = (ideal_min - elev) / ideal_min
        else:  # elev > ideal_max
            # Calculate penalty based on distance from ideal_max
            penalty = (elev - ideal_max) / ideal_min
        return max(0, 1 - penalty)  # Ensure score doesn't drop below 0

# Apply the scoring function
data['Elevation Score'] = elevation.apply(elevation_score)

# Step 4: Min-max scaling for 'AIRD' column
def min_max_scaling(series):
    min_val = series.min()
    max_val = series.max()
    scaled = (series - min_val) / (max_val - min_val)
    return scaled

# Apply the scaling to 'AIRD'
scaled_air_days = min_max_scaling(data['AIRD'])

# Step 5: Calculate the final score
data['Final Score'] = scaled_air_days + data['Elevation Score']

# Optional: Get the top 10 places based on the final score
top_10_places = data.nlargest(10, 'Final Score')[['STATION NAME', 'STATE', 'Final Score', 'ELEVATION (m)', 'AIRD']]

# Print the results to console
print(top_10_places)

# Save the entire DataFrame to a CSV file
output_file_path = 'scaled_results.csv'  # Specify the output file path
data.to_csv(output_file_path, index=False)  # Save without the index

print(f"Results saved to {output_file_path}")
```

```python
import pandas as pd

# Load the data with 'AIRD', 'ELEVATION', and 'distance' columns
file_path = 'Distance_to_Mountains.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Step 1: Extract the relevant columns
elevation = data['ELEVATION']
distance = data['distance']

# Step 2: Define the ideal elevation range
ideal_min = 1500
ideal_max = 3000

# Step 3: Calculate scores for the elevation
def elevation_score(elev):
    if ideal_min <= elev <= ideal_max:
        return 1  # Maximum score for ideal range
    else:
        # Linear drop-off scoring outside the ideal range
        if elev < ideal_min:
            penalty = (ideal_min - elev) / ideal_min
        else:  # elev > ideal_max
            penalty = (elev - ideal_max) / ideal_min
        return max(0, 1 - penalty)

# Apply the scoring function
data['Elevation Score'] = elevation.apply(elevation_score)

# Step 4: Min-max scaling for 'AIRD' column
def min_max_scaling(series):
    min_val = series.min()
    max_val = series.max()
    scaled = (series - min_val) / (max_val - min_val)
    return scaled

# Apply scaling to 'AIRD'
data['AIRD Score'] = min_max_scaling(data['AIRD'])

# Step 5: Inverse min-max scaling for 'distance' (lower values better)
def inverse_min_max_scaling(series):
    min_val = series.min()
    max_val = series.max()
    scaled = (max_val - series) / (max_val - min_val)
    return scaled

# Apply the inverse scaling to 'distance'
data['Distance Score'] = inverse_min_max_scaling(distance)

# Step 6: Calculate the final score
# You can weight each factor if desired by multiplying each by a coefficient
data['Final Score'] = data['AIRD Score'] + data['Elevation Score'] + data['Distance Score']

# Step 7: Reorder and rename columns
data = data[['ICAO', 'STATION_NA', 'COUNTRY', 'LATITUDE', 'LONGITUDE', 'ELEVATION', 'AIRD', 'distance', 
             'Elevation Score', 'AIRD Score', 'Distance Score', 'Final Score']]
data.columns = ['ICAO', 'Station Name', 'Country', 'Lat', 'Long', 'Elevation', 
                'AIRD', 'Distance to Mountains', 'Elevation Score', 
                'AIRD Score', 'Distance Score', 'Final Score']

# Optional: Get the top 10 places based on the final score
top_10_places = data.nlargest(10, 'Final Score')

# Print the top 10 results to console
print(top_10_places)

# Save the entire DataFrame to a CSV file
output_file_path = 'scaled_results.csv'  # Specify the output file path
data.to_csv(output_file_path, index=False)  # Save without the index

print(f"Results saved to {output_file_path}")
```

```python
import pandas as pd

# Load the data with 'AIRD', 'ELEVATION', and 'distance' columns
file_path = 'Distance_to_Mountains.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Convert distance from meters to kilometers
data['distance'] = data['distance'] / 1000

# Step 1: Define the ideal elevation range and calculate elevation score
ideal_min = 1500
ideal_max = 3000

def elevation_score(elev):
    if ideal_min <= elev <= ideal_max:
        return 1  # Maximum score for ideal range
    else:
        penalty = (ideal_min - elev) / ideal_min if elev < ideal_min else (elev - ideal_max) / ideal_min
        return max(0, 1 - penalty)

# Apply elevation score
data['Elevation Score'] = data['ELEVATION'].apply(elevation_score)

# Step 2: Min-max scaling for 'AIRD' column (higher AIRD values better)
def min_max_scaling(series):
    min_val = series.min()
    max_val = series.max()
    return (series - min_val) / (max_val - min_val)

data['AIRD Score'] = min_max_scaling(data['AIRD'])

# Step 3: Inverse min-max scaling for 'distance' (lower distance values better)
def inverse_min_max_scaling(series):
    min_val = series.min()
    max_val = series.max()
    return (max_val - series) / (max_val - min_val)

data['Distance Score'] = inverse_min_max_scaling(data['distance'])

# Step 4: Calculate the final score
data['Final Score'] = data['AIRD Score'] + data['Elevation Score'] + data['Distance Score']

# Step 5: Organize columns in the specified order and rename them
output_data = data[['ICAO', 'STATION_NA', 'COUNTRY', 'LATITUDE', 'LONGITUDE', 'ELEVATION', 'AIRD', 'Distance to Mountains (km', 
                    'Elevation Score', 'AIRD Score', 'Distance Score', 'Final Score']]
output_data.columns = ['ICAO', 'Station Name', 'Country', 'Latitude', 'Longitude', 'Elevation (m)', 
                       'AIRD', 'Distance to Mountains (km)', 'Elevation Score', 'AIRD Score', 
                       'Distance Score', 'Final Score']

# Optional: Display the top 10 places based on the final score
top_10_places = output_data.nlargest(10, 'Final Score')
print(top_10_places)

# Save the cleaned DataFrame to a CSV file
output_file_path = 'cleaned_scaled_results.csv'
output_data.to_csv(output_file_path, index=False)

print(f"Cleaned results saved to {output_file_path}")
```
